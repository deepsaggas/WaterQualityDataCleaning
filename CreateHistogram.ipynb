```python
import pandas as pd
import pandas_profiling
import numpy as np
import time
import glob
import csv
```

## Get all measurement tables

```python
allFiles = glob.glob("SITES_DATA/*/WB_CHUNKS/*/SITES/*.csv")
```

## Calculate frequencies of each unit grouped by characteristic name

```python
dict = {}
columnName = 'ResultMeasure/MeasureUnitCode'
groupByColumnName = 'CharacteristicName'
# for each measurement file/table
for file_ in allFiles:
    df = pd.read_csv(file_, encoding='UTF-8')
    nancount = 0
    # if unit column exists in table
    if columnName in df.columns:
        # for each measuremnt in table update the freq of unit for that particular property
        for (idx, row) in df.iterrows():
            x = row.loc[groupByColumnName];
            unit = str(row.loc[columnName]);
            if(x=="nan"):
                nancount += 1;
            if x not in dict:
                dict[x] = {};
            unitDict = dict[x];
            if unit in unitDict:
                unitDict[unit] = unitDict[unit] + 1;
            else:
                unitDict[unit] = 1;
    else:
        print(columnName + " missing in "+file_)
    if nancount > 0:
        print(str(nancount) + " nan values in " + file_)
print(dict)
```

```python
## Print out the stats to a file group_histogram
f = csv.writer(open("group_histogram.csv", "w"))
f.writerow([groupByColumnName, columnName , "Frequency"])
for i in dict:
    for j in dict[i]:
        f.writerow([str(i), str(j), str(dict[i][j])])
```
